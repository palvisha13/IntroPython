```
#instructor provided dictionary
inventory = {
    'fruit': {
        'apple': 250,
        'pear': 500,
        'orange': 10
    },
    'vegetable': {
        'cucumber': 300,
        'broccoli': 100
    }
}
def menu():
  menu = """Program to change inventory. Type in corresponding number.
  1. Add Category.
  2. Delete Category
  3. Add item 
  4. Delete Item
  5. Change Item
  6. Print Inventory
  0. Exit"""
  return menu

#function to add inventory category
def addCat():
  #user specified category to add 
  add_cat = (input("Add category to inventory. What would you like to call your new category?").strip()).lower()
  if add_cat not in inventory.keys():
    inventory[add_cat]= {}
  elif add_cat in inventory.keys():
    print("That category is already in the inventory. Category replaced by itself.")
  return inventory

  #function to delete category from inventory 
def delCat():
  #user specified category to delete 
  del_cat = (input("Delete category and its items from inventory. Which category would you like to delete?").strip()).lower()
  if del_cat not in inventory.keys():
    print("That category is not in inventory. No category deleted.")
  elif del_cat in inventory.keys():
    del inventory[del_cat]
  return inventory 

#function to add item to category
def addItem():
 #user specified category and item to add to category
  cat_add = (input("What category would you like to add to?").strip()).lower()
  add_item = (input("What item would you like to add to the category?").strip()).lower()
  if cat_add not in inventory.keys():
    print("The category is not in inventory.")
  elif cat_add in inventory.keys():
    inventory[cat_add][add_item]= {}
  return inventory

#function to delete item from inventory 
def delItem():
  #user specified category and item
  cat_del = ((input("What category would you like to delete the item from?")).strip()).lower()
  del_item = ((input("What item would you like to delete?")).strip()).lower()
  if cat_del not in inventory.keys():
    print("The category is not in inventory.")
  elif cat_del in inventory.keys():
    if inventory[cat_del]=={}:
      print("No inventory to delete.")
    else:
      del inventory[cat_del][del_item]
  return inventory 

#function to add value to item
def addValue():
  #user specified category, item, and value
  cat = ((input("What category is your item in?")).strip()).lower()
  item_add= ((input("What item would you like to add or update values for?")).strip()).lower()
  value = input("What numerical value would you like to add?")
  try:
    value =int(value)
    if cat not in inventory.keys():
      print("That category does not exist.")
    elif cat in inventory.keys():
      if item_add not in inventory[cat]:
        print("That item does not exist")
      elif item_add in inventory[cat]:
        inventory[cat][item_add] = value
      
  except:
    print("Value must be an integer value.")


#function to print inventory 
def printInv():
  print(inventory)


```

```
print(menu())
menu_option = (input("Choose an option from the menu: ")).strip()
while True:
  def main():
    if menu_option =='1':
      addInv()
    elif menu_option == '2':
      delInv()
    elif menu_option == '3':
      addItem()
    elif menu_option =='4':
      delItem()
    elif menu_option =='5':
      addValue()
    elif menu_option =='6':
      printInv()
    else:
      print("Choose an option from the menu.")
  main()
  menu_option =(input("Choose an option from the menu: ").lower()).strip()
  if menu_option == '0' or menu_option == 'exit':
    break





```
